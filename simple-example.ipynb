```python
import pandas as pd

train =  pd.read_csv('./titanic/train.csv')
test = pd.read_csv('./titanic/test.csv')
```

```python

# delete unnecessary feature from dataset
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)

train.head(5)
```

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
features_drop = ['Ticket', 'SibSp', 'Parch', 'Embarked', 'Cabin', 'Sex', 'Age']
train = train.drop(features_drop, axis=1)
test = test.drop(features_drop, axis=1)
train = train.drop(['PassengerId'], axis=1)
test = test.drop(['PassengerId'], axis=1)
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']

train_data.shape, target.shape
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
#SVC
clf = SVC(C = 30, gamma = 0.01)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf =  SVC(C = 30, gamma = 0.01)
clf.fit(train_data, target)
```

```python
test_data = pd.read_csv('./titanic/test.csv').copy().drop("PassengerId", axis=1).copy()
test = pd.read_csv('./titanic/test.csv')
features_drop2 = ['Ticket', 'SibSp', 'Parch', 'Embarked', 'Cabin', 'Sex', 'Age', 'Name']
test_data = test_data.drop(features_drop2, axis=1)
test_data.fillna(-99999, inplace=True)
prediction = clf.predict(test_data)

submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": prediction
    })

submission
```
